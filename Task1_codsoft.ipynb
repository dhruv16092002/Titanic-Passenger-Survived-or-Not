#  Model Deployment


```python
# improting spliting model form model selection
from sklearn.model_selection import train_test_split
```

```python
# importing preprocessing librar form skelarn
from sklearn import preprocessing
# importing lable endoered form preprocessing
label_encoder = preprocessing.LabelEncoder()
 
# lableing in dataset of sex column
dataset['Sex']= label_encoder.fit_transform(dataset['Sex'])

print('1 for male and 0 for female')
dataset['Sex'].unique()

```

```python
dataset
```

```python
# splitting dataset into x_train, y_train, x_test and y_test
x = dataset[['Pclass','Sex']]
y = dataset['Survived']
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
# Importing logisticRegression form skelarn linear_model
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
# Prediction of x_test dataset
print(log.predict(x_test))
```

```python
import warnings

# Suppress the specific warning
warnings.filterwarnings("ignore", category=UserWarning, message="X does not have valid feature names, but LogisticRegression was fitted with feature names")


user_class = int(input())
user_gender = input()
u_g = 0
if user_gender == 'male':
    u_g = 1
elif user_gender == 'female':
    u_g = 0
else:
    print('Please enter gender correctly')
res = log.predict([[user_class,u_g]])

if (res == 0):
    print('The person is Not Survived')
else:
    print('the person is Survived')


# Reset warning filters to their default state (optional)
warnings.resetwarnings()
```
